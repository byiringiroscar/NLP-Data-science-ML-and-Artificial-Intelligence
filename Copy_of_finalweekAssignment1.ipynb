```python
import pandas as pd
import re
import string
df = pd.read_csv('/content/drive/MyDrive/NLP Fellowship/huzalab_doc/week2/Eng_Kin-Paralleldata (1).csv')
dataset = df[['English','Kinyarwanda']]
```

```python
df_back = dataset
df_back= df_back.applymap(lambda s:s.lower() if type(s) == str else s)
# df_back_p = df_back.applymap(lambda s: s.translate(str.maketrans('', '', string.punctuation)))
df_back.head(5)
```

```python
all_email = []
all_link = []
regexE = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
regexL = r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+"
def check(data):
  new_word = data.split()
  for n in new_word:
    if(re.fullmatch(regexE, n)):
        all_email.append(n)
    elif(re.fullmatch(regexL, n)):
      all_link.append(n)
df_back_new = dataset.applymap(lambda s: check(s))
all_email,all_link
# len(dataset)
```

```python
def sub_change_email(data):
    data = re.sub(regexE, "", data)
    return data
def sub_change_link(data):
    data = re.sub(regexL, "", data)
    return data

df_back_final = dataset.applymap(lambda s: sub_change_email(s))
df_back_final_new = df_back_final.applymap(lambda s: sub_change_link(s))
df_back_final_new.head(5)
```

```python
all_email_link = []
all_link_link = []
regexE = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
regexL = r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+"
def check_new(data):
  new_word = data.split()
  for n in new_word:
    if(re.fullmatch(regexE, n)):
        all_email_link.append(n)
    elif(re.fullmatch(regexL, n)):
      all_link_link.append(n)
df_back_new = df_back_final_new.applymap(lambda s: check_new(s))
all_email_link,all_link_link
df_back_final_new.head(5)
```

```python
df_back_p = df_back_final_new.applymap(lambda s: s.translate(str.maketrans('', '', string.punctuation)))
len(df_back_p)
df_back_p.to_csv('df_back_final_new.csv')
```
