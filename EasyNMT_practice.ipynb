```python
from easynmt import EasyNMT
```

```python
model = EasyNMT("opus-mt")
```

```python
sentences=['Jesus is alive',
           'Hands on projects is our priority',
           "I don't have anything",
           "It is  what is!:legend Balise said",
           "I don't have anything to give you baby"]
```

```python
translation = model.translate(sentences, target_lang="fr")
translation
```

```python
import pandas as pd
```

```python
dataset = pd.DataFrame()
```

```python
dataset["french"] = translation
dataset['Engish'] = model.translate(dataset["french"], source_lang='fr', target_lang="en")
dataset
```
