```python
import json
def post_character(image, title_head,  final_description):
  try:
    description = title_head + " " + final_description
    url = 'https://jormungandr.wizewerx.tech/eat'
    headers = {
        'Authorization': 'p7RF3TwoFtaQ9ymoFMS3ZdYfSdc7hEoYedXxZ1L5l7vSXf59ntjvZTymoF5rfpy8RSpD4IErZt2ykSv64Xbu86l7',
        'Content-Type': 'application/json'
    }

    data = {
        "topic": "fuk",
        "description": description,
        "payload": {},
        "files": image
    }
    json_data = json.dumps(data)
    response = requests.post(url, headers=headers, data=json_data)
    if response.status_code == 200:
        print("done")
    else:
        print("not done")
  except:
    pass
```

```python
def get_link_details(url):
  content_main = requests.get(url).content
  page = BeautifulSoup(content_main, 'html.parser')
  images_div = page.find('div', class_='product-gallery__thumbnail-list')
  images = images_div.find_all('a')
  final_images = []
  for img in images:
    link_image = img['href']
    full_link = "https://" + link_image[2:]
    final_images.append(full_link)
  title_element = page.find('h1', class_='product-meta__title')
  title = title_element.text
  description_div = page.find('div', class_='rte text--pull')
  description = description_div.text.strip()
  post_character(final_images, title, description)
  # return final_images, title, description
```

```python
def get_links(url):
  content_main = requests.get(url).content
  page = BeautifulSoup(content_main, 'html.parser')
  categorycharacter = page.find_all('a', class_="product-item__image-wrapper")
  links_page = []
  for link in categorycharacter:
    base_link = "https://fifthsun.com"
    full_link = base_link + link['href']
    links_page.append(full_link)
  return links_page
```
