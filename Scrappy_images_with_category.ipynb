```python
import os
import requests
from urllib.parse import urlparse, unquote
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
#Fetch the content from the front page
url_links = 'https://www.next.co.uk/'
content = requests.get(url_links).content
soup = BeautifulSoup(content, 'html.parser')
```

```python
# finding navbar
navbar_category = []
navbar_elements = soup.find('ul', id='snail-trail-container')
for li_element in navbar_elements.find_all('li'):
    a_element = li_element.find('a')
    href = a_element['href']
    navbar_category.append(href)

```

```python
def get_images_links_all(url):
  category = url.split("/")[-1]
  response = requests.get(url)
  soup = BeautifulSoup(response.content, "html.parser")

  image_tags = soup.find_all("img")
  base_url = "https://www.next.co.uk"

  valid_extensions = ('.png', '.jpg', '.jpeg')

  image_urls = [{"urls": urljoin(base_url, img["src"]), "category": category} for img in image_tags if img.get("src") and img["src"].endswith(valid_extensions)]
  return image_urls
```

```python
final_images_link = []
for links in navbar_category:
  final_images_link.extend(get_images_links_all(links))
```

```python
print(final_images_link)
```

```python
def get_filename_and_extension(image_url):
    parsed_url = urlparse(image_url)
    path_segments = parsed_url.path.split('/')
    filename_with_query = path_segments[-1]

    # Remove query parameters from the filename
    filename = filename_with_query.split('?')[0]
    if '.' not in filename:
        filename += '.jpeg'
    filename = unquote(filename)

    return filename
```

```python
# List of image URLs

# Create a folder to save the images

def download_images():
  for index, links in enumerate(final_images_link):
    if(index) < 5:
      response = requests.get(links['urls'])
      if response.status_code == 200:
          image_filename = get_filename_and_extension(links['urls'])
          os.makedirs(f"/content/scrapy_images/{links['category']}", exist_ok=True)
          save_path = os.path.join(f"/content/scrapy_images/{links['category']}", image_filename)

          with open(save_path, "wb") as f:
              f.write(response.content)
download_images()
```
